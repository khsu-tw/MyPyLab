```python
%matplotlib inline
import numpy as np
import numpy.random as random
import matplotlib.pyplot as plt

amplitidue = eval(input('請輸入脈衝訊號震幅：'))
probability = eval(input('請輸入脈衝訊號發生機率(%)'))

t=np.linspace(0,1,200)
x=10*np.cos(2*np.pi*5*t)

noise = np.zeros(x.size)
for i in range (x.size):
    p1=random.uniform (0,1)
    if p1<probability /100:
        p2 = random.uniform(0,1)
        if p2<0.5:
            noise[i]=amplitidue
        else:
            noise[i] = -amplitidue
            
y=x+noise

#畫出原始訊號
plt.figure(1)
plt.plot(t, x)
plt.xlabel('t(sec)')
plt.ylabel('Amplitude')
plt.axis([0,1,-15,15])

#脈衝雜訊訊號
plt.figure(2)
plt.plot(t, noise)
plt.xlabel('t(sec)')
plt.ylabel('Amplitude')

#原始訊號加入脈衝雜訊
plt.figure(3)
plt.plot(t, y)
plt.xlabel('t(sec)')
plt.ylabel('Amplitude')
plt.axis([0,1,-15,15]);
```
