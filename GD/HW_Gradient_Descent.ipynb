# 簡單利用cos(x)做為Loss function做梯度下降測試

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from numpy import arange

def tF(x): 
    return np.cos(x)

def dtF(x): 
    return np.sin(-x)



def GD(x_start, df, epochs, lr): #（起始點,一階函數微分,通過NN運算次數,學習率）
    xEPO = np.zeros(epochs+1)    
    x = x_start    
    xEPO[0] = x    
    for j in range(epochs):         
        dx = df(x)        
        k = - dx * lr        
        x = x + k        
        xEPO[j+1] = x    
    return xEPO

x_start = 5    
epochs = 100
lr_h = 3
#lr_l = 0.2
x = GD(x_start, dtF, epochs, lr=lr_h) 
print (x)

color = 'r'    
t = arange(-6.0, 6.0, 0.01)
plt.plot(t, tF(t), c='b')
plt.plot(x, tF(x), c=color, label='lr={}'.format(lr_h))    
plt.scatter(x, tF(x), c=color); 
```

### 上圖結果可發現在執行梯度下降的第二次循環後就開始出現左右循環跳動，而且一直無法找到最佳解。

### 下面僅針對學習率調降為原來(lr=3)的1/10倍

```python
lr_l = 0.3
x1 = GD(x_start, dtF, epochs, lr=lr_l) 
print (x1)

color = 'r'    
t = arange(-6.0, 6.0, 0.01)
plt.plot(t, tF(t), c='b')
plt.plot(x1, tF(x1), c=color, label='lr={}'.format(lr_l))    
plt.scatter(x1, tF(x1), c=color); 
```

### 上圖再修改學習率後，到了第55次就可找到最佳解3.14159265

### 下面再將學習率調降為原來(lr=3)的1/100倍

```python
#將learning rate降為原來的1/100倍
lr_ll = 0.03
x2 = GD(x_start, dtF, epochs, lr=lr_ll) 
print (x2)

color = 'r'    
t = arange(-6.0, 6.0, 0.01)
plt.plot(t, tF(t), c='b')
plt.plot(x2, tF(x2), c=color, label='lr={}'.format(lr_ll))    
plt.scatter(x2, tF(x2), c=color); 
```

### 若以第二張圖訓練結果為參考，也就是3.1415926為最佳解，最後這張圖在100次循環後，仍無法找到最佳解

## 綜合上面三張圖，學習率設定適當與否對找到最佳解有很重要的影響
