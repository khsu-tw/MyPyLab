```python
af2007=gapminder[(gapminder['continent']=='Africa') & (gapminder['year']==2007)] #指定變數af2007為非洲各國＋2007年資料
```

```python
af2007[['lifeExp']].mean() #看看2007年非洲國家人口的平均壽命
```

```python
af2007[['lifeExp']].min() #看看2007年非洲國家人口的最低壽命
```

```python
gapminder[(gapminder['continent'] == 'Africa')& (gapminder['lifeExp']==39.613)] #找出2007年非洲平均壽命最低的國家，原來是史瓦濟蘭！
```

```python
#先畫個非洲的散佈圖，結果意外的看到有一個點非常的低，所以2007年的39.613歲不是最低的壽命值。
#透過這張圖我想要找出是那一年、幾歲、哪一個非洲國家
gapminder_af = gapminder[gapminder['continent'] == 'Africa']
gapminder_af[['year', 'lifeExp']].plot(kind = 'scatter', x = 'year', y = 'lifeExp', title = 'LifeExp vs. Year in Africa');
```
