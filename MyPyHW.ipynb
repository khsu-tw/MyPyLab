```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
from PIL import Image

# 開啟一個圖檔，圖檔與主程式放於同一個目錄
im = Image.open("Mario.jpg")
# 取得圖檔資訊:
w, h = im.size
print('原始圖檔尺寸: %sx%s' % (w, h))
#print('Original image size: %sx%s' % (width, height))

# 縮放圖檔比例
def resize(width,height):
    print('Resize image to: %sx%s' % (width, height))
    img = im.resize((width, height), Image.BILINEAR)
    plt.imshow(img);
interact (resize, width=(w,(w*2),50), height=(h,(h*2),50))

# 將縮放後的圖檔從新存成JPG:
#im.save('thumbnail.jpg', 'jpeg')
```

```python
%matplotlib inline
import numpy as np
import numpy.random as random
import matplotlib.pyplot as plt

amplitidue = eval(input('請輸入脈衝訊號震幅：'))
probability = eval(input('請輸入脈衝訊號發生機率(%)'))

t=np.linspace(0,1,200)
x=10*np.cos(2*np.pi*5*t)

noise = np.zeros(x.size)
for i in range (x.size):
    p1=random.uniform (0,1)
    if p1<probability /100:
        p2 = random.uniform(0,1)
        if p2<0.5:
            noise[i]=amplitidue
        else:
            noise[i] = -amplitidue
            
y=x+noise

#畫出原始訊號
plt.figure(1)
plt.plot(t, x)
plt.xlabel('t(sec)')
plt.ylabel('Amplitude')
plt.axis([0,1,-15,15])

#脈衝雜訊訊號
plt.figure(2)
plt.plot(t, noise)
plt.xlabel('t(sec)')
plt.ylabel('Amplitude')

#原始訊號加入脈衝雜訊
plt.figure(3)
plt.plot(t, y)
plt.xlabel('t(sec)')
plt.ylabel('Amplitude')
plt.axis([0,1,-15,15]);
```
